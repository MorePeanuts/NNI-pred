```python
from nni_pred.data import VariableGroups, MergedTabularDataset
from nni_pred.transformers import get_preprocessing_pipeline
import sklearn

# sklearn.set_config(transform_output='pandas')
```

```python
data = MergedTabularDataset()
X, y_dict, group = data.prepare_data()

pipeline = get_preprocessing_pipeline('linear')
X = pipeline.named_steps['keeper'].fit_transform(X)
print(type(X), X.isna().sum().sum())

X = pipeline.named_steps['encoder'].fit_transform(X)
print(type(X), X.isna().sum().sum())
print(X.columns)

X = pipeline.named_steps['skew_shift'].fit_transform(X)
print(type(X), X.isna().sum().sum())
print(X.head(5))

X = pipeline.named_steps['column_transformer'].fit_transform(X)
print(type(X), X.isna().sum().sum())
if X.isna().any().any():
    print(X.columns[X.isna().any()].tolist())
```
