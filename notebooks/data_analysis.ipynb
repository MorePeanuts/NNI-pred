# Data Analysis

This is a notebook used for data analysis based on `merged_data.csv` from spatial feature engineering script `sfe_merge.py`

```python
import pandas as pd
import numpy as np

data = pd.read_csv('../datasets/merged_data.csv')
print(data.info())
print(data.columns)
```

Analyze the skewness of the data to identify the characteristics of a long-tailed distribution.

```python
numeric_cols = data.select_dtypes(include=[np.number]).columns.tolist()
exclude_cols = ['ID', 'Season', 'Lon', 'Lat', 'Landuse', 'Soil_landuse']
candidates = [c for c in numeric_cols if c not in exclude_cols]
skew_limit = 0.75

skewness = data[candidates].skew()
high_skew_cols = skewness[skewness > skew_limit].index

print(f"Detected {len(high_skew_cols)} right-skewed variables, performing Log1p transformation...")
# for col in high_skew_cols:
#     if (data[col] >= 0).all():
#         data[col] = np.log1p(data[col])
```
